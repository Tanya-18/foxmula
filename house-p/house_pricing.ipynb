```python
dfclean = df[df['Outliers']==0]
print(df.shape)
print(dfclean.shape)
```

```python
dfclean = dfclean.drop(columns=['count','Outliers'])
dfclean.shape
```

```python
dfclean = dfclean.drop(columns=['date', 'street', 'city', 'statezip', 'country'])
dfclean.shape
```

```python
x = dfclean.drop(columns='price')
y = dfclean['price']

print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.2)

print(xtrain.shape)
print(xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```

```python
import statsmodels.api as sm

model = sm.OLS(ytrain, xtrain).fit()
```

```python
print(model.summary())
```

```python
pred = model.predict(xtest)
data = list(zip(ytest, pred))
```

```python
comp = pd.DataFrame(data, columns=['Actual', 'Predicted'])
comp.head()
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt

rmse = sqrt(mean_squared_error(ytest, pred))
print(rmse)
```
